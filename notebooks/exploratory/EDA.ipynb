```python
categoricas = [
    'modalidade', 'situacao_candidato', 'informacoes_pessoais_sexo',
    'informacoes_pessoais_estado_civil', 'informacoes_profissionais_nivel_profissional',
    'formacao_e_idiomas_nivel_academico'
]

for col in categoricas:
    if col in df.columns:
        print(f"\nDistribuição de {col}:")
        print(df[col].value_counts(dropna=False))
```

```python
numericas = [
    'valor_venda', 'informacoes_profissionais_remuneracao'
]

for col in numericas:
    if col in df.columns:
        print(f"\nAnálise de {col}:")
        print(df[col].describe())
```

```python
datas = [
    'data_candidatura', 'ultima_atualizacao',
    'informacoes_pessoais_data_nascimento', 'cargo_atual_data_admissao'
]

for col in datas:
    if col in df.columns:
        print(f"\n{col} - intervalo de datas:")
        print(df[col].min(), "→", df[col].max())
```

```python
if 'data_candidatura' in df.columns:
    df['data_candidatura'] = pd.to_datetime(df['data_candidatura'], errors='coerce')
    df['ano_mes_candidatura'] = df['data_candidatura'].dt.to_period('M')
    
    # Contar as candidaturas por ano-mês e exibir os valores
    candidaturas_por_mes = df['ano_mes_candidatura'].value_counts().sort_index()
    print(candidaturas_por_mes)
```
